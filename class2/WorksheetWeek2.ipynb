# Encodings and Marks

Now that we have learned about all kinds of marks and options with their design, we will practice by exploring our options. First, import your libraries and load the data-- today, we will be working on the `seattle_weather` dataset from `vega_datasets`.

```python
from vega_datasets import data
weather = data.seattle_weather()
weather.head()
```

### Plot 1

Make a scatterplot of precipitation over time. Scale the precipitation axis so that values between 0-20 are discernable. Then, choose another encoding channel to represent what type of weather it was at that time. 

**Why did you choose that encoding?**

```python
weather['Year'] = weather['date'].dt.year
chart = alt.Chart(weather)
alt.Chart(weather).mark_point().encode(
    alt.X('date:T'),
    alt.Y('precipitation:Q', scale=alt.Scale(zero=False)),
    alt.Color('weather:N', title='Weather Type'),
    tooltip=['date', 'precipitation', 'weather']
    )
```

In the chart above date(Temporal) is used as a x encoding channel which sets a mark's horizontal position (x-coordinate). precipitation (quantitative) is used as a y encoding channel which sets a mark's vertical position (y-coordinate)
Encoded the weather type field using the color channel and a nominal (N) data type, resulting in a distinct hue for each weather type making it easy to distinguish weather types visually and also accessible for color-blind users. 
Tooltip encoding channel is added inspecting exact values and weather types.

### Plot 2
Show the same variables as in plot 1, but now use a line mark. Use interpolation to smooth the choppy lines and change the aspect ratio/size so the time variable is more visible. Save this plot as a base chart, but don't print/output it. 

Then, plot of maximum precipitation for each individual month between 2012 and 2015 (i.e. Jan 2012, Feb 2012, ..., Nov 2015, Dec 2015) made from your base plot. [The documentation for time transform functions](https://altair-viz.github.io/user_guide/transform/timeunit.html#user-guide-timeunit-transform) may be helpful here. 

Also, plot the minimum, average, and maximum precipitation for each individual month between 2012 and 2015 (i.e. Jan 2012, Feb 2012, ..., Nov 2015, Dec 2015). 

You can display them vertically stacked using &. 

**What do you notice about these three plots? How could these be improved? (You do not need to make the improvement, just point out what is misleading about these plots).**

```python
base = alt.Chart(weather).mark_line(interpolate='monotone').encode(
    alt.X('yearmonth(date):T'),
    alt.Y('precipitation:Q'),
    alt.Color('weather:N')    
).properties(width=800, height=250)

base1 = base.encode( 
     y=alt.Y('min(precipitation):Q', title='Min Precipitation'),
    ).properties(title='Minimum Monthly Precipitation (20122015)')
base2 = base.encode( 
    y=alt.Y('mean(precipitation):Q', title='Average Precipitation'),
    ).properties(title='Average Monthly Precipitation (20122015)')
base3 = base.encode( 
    y=alt.Y('max(precipitation):Q', title='Max Precipitation'),
    ).properties(title='Maximum Monthly Precipitation (20122015)')

base1 & base2 & base3
```

-It's hard to visualize the exact values(max/min/mean) for a particular month in all the three graphs in this case bar chart would be more helpful for interpretation. Moreover Different Y-axis scales across plots make cross-comparison misleading. Aligning them or using a shared scale would provide more accurate visual comparison. interpolate='monotone' on monthly aggregates in-between values and can distort extrema.

### Plot 3

Make a series of bar charts showing the frequency of each type of weather over the calendar months. 
1. Make one stacked (the default), called stacked
2. Make one that transforms the total records so they show as a percentage (i.e. have a scale from 0 to 1), called unstacked
3. Make an area plot of the same information with the layers only 10% opaque and not stacked, called area
  
Resizing them as needed to fit in the page and show these three plots side-by-side with stacked | unstacked | areas

Note: The second version of the plot only requires one small additional parameter. Look in the documentation to find your options, but don't transform the data outside of the plot-- Altair is smart enough to do this for us!

```python

# 1) Stacked bar chart (default stacking)
stacked = (
    alt.Chart(weather).mark_bar().encode(
        alt.X('month(date):O',title='Month'),
        alt.Y('count():Q', title='Frequency'),
        alt.Color('weather:N', title='Weather'),
        tooltip=[alt.Tooltip('month(date):O', title='Month'), 'weather:N', alt.Tooltip('count():Q', title='Freq')]
    )
    .properties(title='stacked', width=260, height=240)
)
unstacked =(
    alt.Chart(weather).mark_bar().encode(
        alt.X('month(date):O',title='Month'),
        alt.Y('count():Q', stack='normalize', title='Proportion'),
        alt.Color('weather:N', title='Weather'),
        tooltip=[alt.Tooltip('month(date):O', title='Month'), 'weather:N', alt.Tooltip('count():Q', title='Freq')]
    )
    .properties(title='unstacked', width=260, height=240)
)

areas = (
    alt.Chart(weather).mark_area(opacity=0.10).encode(
        alt.X('month(date):O',title='Month'),
        alt.Y('count():Q', stack=None, title='Frequency'),
        alt.Color('weather:N', title='Weather'),
        tooltip=[alt.Tooltip('month(date):O', title='Month'), 'weather:N', alt.Tooltip('count():Q', title='Freq')]
    )
    .properties(title='areas', width=260, height=240)
)

stacked| unstacked | areas
```

### Plot 4

Make a tick plot showing the minimum temperature for each month between 2012 and 2015. You can use `yearmonth(date)` as the x position to make sure the ticks are short enough so they don't overlap with other months' data, and change the opacity so you can see when ticks overlap.  

Then, layer on a second set of ticks, all in red, that shows the maximum temperature for the same month by filling in the what I have started in the cell. You should have a single plot with both red and blue ticks. 

```python
blue = (
    alt.Chart(weather)
    .mark_tick(orient='horizontal', color='steelblue', opacity=0.7, thickness=2, size=20)
    .encode(
        x=alt.X('yearmonth(date):T', title='Month', axis=alt.Axis(format='%b %Y', labelAngle=0)),
        y=alt.Y('min(temp_min):Q', title='Temperature (C)'),
        tooltip=[alt.Tooltip('yearmonth(date):T', title='Month'),
                 alt.Tooltip('min(temp_min):Q', title='Min Temp (C)')]
    )
    .properties(width=900, height=320, title='Monthly Min (blue) & Max (red) Temperatures, 20122015')
)
blue + blue.mark_tick(orient='horizontal', color='red', opacity=0.7, thickness=2, size=20).encode(
    y=alt.Y('max(temp_max):Q', title='Temperature (C)'),
    tooltip=[alt.Tooltip('yearmonth(date):T', title='Month'),
             alt.Tooltip('max(temp_max):Q', title='Max Temp (C)')]
)
```

### Plot 5

Finally, combine (almost) all your data together into one plot! Make a filled scatterplot of the minimum temperature vs maximum temperature. Color your points by the weather conditions, size them by the wind speed, and turn down the opacity so you can see the layered points better. Then, facet by weather condition into columns. Change the size of your plots so they can be seen on one page (it's ok if they are fairly small!).      

```python
base_scatter = (
    alt.Chart(weather)
    .mark_point(filled=True, opacity=0.35)
    .encode(
        alt.X('temp_min:Q', title='Min Temperature (C)'),
        alt.Y('temp_max:Q', title='Max Temperature (C)'),
        alt.Color('weather:N', title='Weather'),
        alt.Size('wind:Q', title='Wind Speed'),
        alt.Column('weather:N'),
        tooltip=[
            alt.Tooltip('date:T', title='Date'),
            alt.Tooltip('temp_min:Q', title='Min Temp (C)'),
            alt.Tooltip('temp_max:Q', title='Max Temp (C)'),
            alt.Tooltip('wind:Q', title='Wind'),
            'weather:N'
        ]
    )
    .properties(width=150, height=150)
)
base_scatter
```
